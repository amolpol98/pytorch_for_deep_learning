# Tensor operations and gradients

```python
x = torch.tensor(3.)
w = torch.tensor(4., requires_grad=True)
b = torch.tensor(5., requires_grad=True)
```

```python
y = w*x + b
y
```

```python
# Compute derivates
y.backward()
```

```python
print('dy/dx:', x.grad)
print('dy/dw:', w.grad)
print('dy/db:', b.grad)
```
